```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('breast_cancer_company.csv')
```

```python
data = data.drop(columns='Unnamed: 32', axis=1)
```

```python
data.head()
```

```python
X = data.drop(columns=['diagnosis','id'] , axis=1)
y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split as TTS
```

```python
X_train,X_test,y_train,y_test = TTS(X,y,test_size=0.2,random_state=42)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler as SC
sc = SC()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.linear_model import LinearRegression as LRs
from sklearn.tree import DecisionTreeClassifier as DTC
```

```python
models = [LR(max_iter=10000, random_state=42), KNN(n_neighbors=5), RF(criterion='entropy', random_state=42), DTC(criterion='gini', random_state=42)]
```

```python
from sklearn.metrics import accuracy_score as AS
from sklearn.metrics import confusion_matrix as CM
from sklearn.metrics import classification_report as CR
```

```python
# model training and testing.
def model_training(models):
  for model in models:
      model.fit(X_train, y_train)
      y_pred = model.predict(X_test)
      print(f"\n{model} accuracy: {AS(y_test, y_pred)}")
      print("\nConfusion Matrix: ")
      print(CM(y_test, y_pred))
      print("\nClassification Report: ")
      print(CR(y_test, y_pred))
      print("\n\n")
```

```python
model_training(models)
```
